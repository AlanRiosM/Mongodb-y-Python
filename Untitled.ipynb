```python
import pymongo

# Conectarse a la base de datos
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["mi_base_de_datos"]

# Crear
def insertar_libro(datos_libro):
    col = db["libros"]
    resultado = col.insert_one(datos_libro)
    print(f"Libro insertado con el ID: {resultado.inserted_id}")

# Leer
def consultar_libros(filtro=None):
    col = db["libros"]
    libros = col.find(filtro) if filtro else col.find()
    for libro in libros:
        print(libro)

# Actualizar
def actualizar_libro(filtro, nuevos_datos):
    col = db["libros"]
    resultado = col.update_one(filtro, {"$set": nuevos_datos})
    print(f"Se actualizaron {resultado.modified_count} libros")

# Borrar
def borrar_libro(filtro):
    col = db["libros"]
    resultado = col.delete_one(filtro)
    print(f"Se eliminó {resultado.deleted_count} libro")
```

```python
nuevo_libro = {
        "Book": "Ejemplo de Libro",
        "Author(s)": "Autor1",
        "Original language": "es",
        "First published": 2022,
        "Approximate sales in millions": 2.5,
        "Genre": ["Ficción", "Misterio"]
}

insertar_libro(nuevo_libro)
consultar_libros()
```

```python
actualizar_libro({"Book": "Ejemplo de Libro"}, {"Approximate sales in millions": 3.0})
consultar_libros()
```

```python
borrar_libro({"Book": "Ejemplo de Libro"})
consultar_libros()
```
